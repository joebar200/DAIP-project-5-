```python
data.drop(['id','customerID'],axis = 1, inplace = True)
```

```python
#Replace all yes/no with 0/1
for i in data.columns:
    for j in range(0,len(data)):
        if data[i][j] == "No phone service":
            data.loc[j,i] = 0
        elif data[i][j] == "Yes":
            data.loc[j,i] = 1
        elif data[i][j] == "No":
            data.loc[j,i] = 0
```

```python
#Replace "No internet service" in internet-related attributes with zero
for i in internetatt:
    for j in range(0,len(data)):
        if data[i][j] == "No internet service":
            data.loc[j,i] = 0
```

```python
#Check if no internet implies no other options related to internet
counter = 0
for i in range(0,len(data)):
        if data['InternetService'][i] == 0:
            counter = counter + data.iloc[i,8:14].sum()
counter
#sum is 0 so it is true
```

```python
data['Fiber optic'] = 0
data['DSL'] = 0
for i in range(0,len(data)):
    if data['InternetService'][i] == 'Fiber optic':
        data.loc[i,'Fiber optic'] = 1
    elif data['InternetService'][i] == 'DSL':
        data.loc[i,'DSL'] = 1
data.drop('InternetService',axis =1, inplace = True)
```

```python
data.head(5)
```

```python
data['Male'] = 0
data['Female'] = 0
for i in range(0,len(data)):
    if data['gender'][i] == 'Male':
        data.loc[i,'Male'] = 1
    elif data['gender'][i] == 'Female':
        data[i,'Female'] = 1
data.drop('gender',axis =1, inplace = True)
```

```python
data['Contract'].unique()
```

```python
data['Two year'] = 0
data['One year'] = 0
for i in range(0,len(data)):
    if data['Contract'][i] == 'One year':
        data.loc[i,'One year'] = 1
    elif data['Contract'][i] == 'Two year':
        data[i,'Two year'] = 1
data.drop('Contract',axis =1, inplace = True)
```

```python
data['PaymentMethod'].unique()
```

```python
data['Credit card (automatic)'] = 0
data['Bank transfer (automatic)'] = 0
data['Electronic check'] = 0

for i in range(0,len(data)):
    if data['PaymentMethod'][i] == 'Credit card (automatic)':
        data.loc[i,'Credit card (automatic)'] = 1
    elif data['PaymentMethod'][i] == 'Bank transfer (automatic)':
        data[i,'Bank transfer (automatic)'] = 1
    elif data['PaymentMethod'][i] == 'Electronic check':
        data[i,'Electronic check'] = 1
data.drop('PaymentMethod',axis =1, inplace = True)
```

```python
column_names = ['Male','Female','SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'Fiber optic', 'DSL',
        'OnlineSecurity','OnlineBackup', 'DeviceProtection', 'TechSupport',
        'StreamingTV','StreamingMovies', 'One year','Two year', 'PaperlessBilling','Credit card (automatic)',
        'Bank transfer (automatic)','Electronic check',
       'MonthlyCharges', 'TotalCharges', 'Churn'] #Rearrange columns
```

```python
data = data.reindex(columns=column_names)
```
